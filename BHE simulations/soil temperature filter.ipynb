```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv(r'~/Desktop/test.csv')

for index, row in df.iterrows():
    if row['MaterialID'] == 36:
        df.drop(index, inplace=True)


plt.plot(df['Depth'], df['Temperature'], marker='o', linestyle='-')
plt.xlabel('Depth (meters)',fontsize=14)
plt.ylabel('Temperature (Celsius)')
plt.title('Depth vs Temperature')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df2 = pd.read_csv(r'~/Desktop/test_end1.csv')

for index, row in df2.iterrows():
    if row['MaterialID'] == 36:
        df2.drop(index, inplace=True)


plt.plot(df2['Depth'], df2['Temperature'], marker='o', linestyle='-')
plt.xlabel('Depth (meters)')
plt.ylabel('Temperature (Celsius)')
plt.title('Depth vs Temperature')
plt.grid(True)
plt.show()
```

```python
plt.plot(df['Temperature'], df['Depth'], marker='o', linestyle='-', color='red', label='305 seconds')
plt.plot(df2['Temperature'], df2['Depth'], marker='o', linestyle='-', color='green', label='6 month')
plt.xlabel('Temperature ( C)',fontsize=18,fontweight='bold')
plt.ylabel('Depth (m)',fontsize=18,fontweight='bold')
plt.title('Temperature vs. Depth',fontsize=22,fontweight='bold')
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.grid(True)
plt.rcParams['figure.figsize'] = [10, 10]
plt.legend(fontsize=20)

plt.savefig('soil_temperature_comparison.png')
plt.show()
```
